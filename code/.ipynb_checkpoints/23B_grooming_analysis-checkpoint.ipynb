```python
#Optimized helper functions to do the calculation above
def get_closest_joint(val_array, pt_cols, joint):
    """
    Finds the closest joint to a given joint in a point cloud.
    Parameters:
    - val_array (numpy.ndarray): The array of feature vectors representing the point cloud.
    - pt_cols (list): The list of joint names corresponding to the feature vectors.
    - joint (str): The name of the joint to find the closest joint to.
    Returns:
    - closest (int): The index of the closest joint in the point cloud.
    - dist (float): The distance between the given joint and the closest joint.
    """
    
    kdt = KDTree(val_array, leaf_size=3, metric='euclidean')

    #Id of our joint of interest
    idx = [ix for ix, i in enumerate(pt_cols) if i == joint][0]
    X = val_array[idx].reshape(1,3) #corresponding feature vector for that joint

    #Query for the top 2 nearest neighbors since the first will be itself
    ds, similar_idxs = kdt.query(X, k=2, return_distance=True)
    closest = similar_idxs[0][1]
    dist = ds[0][1]
    
    return closest, dist

def get_joint_distances(joint, df, pt_cols=[]):
    """
    Calculate the distances between a given joint and the closest joint in a DataFrame.
    Parameters:
    - joint (str): The joint to calculate distances from.
    - df (pandas.DataFrame): The DataFrame containing the joint positions.
    - pt_cols (list, optional): The columns in the DataFrame that contain the joint positions. 
                                If not provided, all columns containing 'pt_position' in their names will be used.
    Returns:
    - distances (tuple): A tuple of distances between the given joint and the closest joint for each row in the DataFrame.
    - closest (tuple): A tuple of the closest joint for each row in the DataFrame.
    """
    
    if len(pt_cols) == 0:
        pt_cols = [i for i in df.columns if 'pt_position' in i]
    vals = df.apply(lambda row: np.vstack(row[pt_cols].values), axis=1)
    df['vals'] = vals
    closest_joints = df.apply(lambda row: get_closest_joint(row['vals'],pt_cols,joint), axis=1)
    closest, distances = zip(*closest_joints)

    return closest, distances

def get_joint_df(df, joint, root_leg, dist_threshold=0.27, interval=60):
    """
    Generate a DataFrame containing information about joint distances and contacts.
    Parameters:
    - df (pandas.DataFrame): The input DataFrame containing the data.
    - joint (str): The name of the joint.
    - root_leg (str): The name of the root leg.
    - dist_threshold (float, optional): The distance threshold for considering a contact. Defaults to 0.27.
    - interval (int, optional): The interval for grouping the data. Defaults to 60.
    Returns:
    - joint_df (pandas.DataFrame): The resulting DataFrame with the following columns:
        - 'root_joint': The name of the joint.
        - 'stim': The stimulus type.
        - 'fnum': The frame number.
        - 'distance': The distance between the joint and the closest point.
        - 'closest_idx': The index of the closest point.
        - 'closest_label': The label of the closest point.
        - 'repnum': The repetition number.
        - 'roundnum': The round number.
        - 'fly': The fly identifier.
        - 'is_root_leg': A boolean indicating if the root leg is present in the closest label.
        - 'is_contact': A boolean indicating if a contact is present based on the distance threshold.
        - 'interval': The interval number based on the frame number.
    """
   
    joint_df = pd.DataFrame()
    pt_cols = [i for i in df.columns if 'pt_position' in i]         
    on_distances, on_closest = get_joint_distances(joint,df, pt_cols=pt_cols)
    on_closest_labels = [pt_cols[i] for i in on_closest]
   
    joint_df['root_joint'] = [joint] * len(on_distances)
    joint_df['stim'] = ['on'] * len(on_distances)
    joint_df['fnum'] = df.fnum.tolist()
    joint_df['distance'] = on_distances
    joint_df['closest_idx'] = on_closest
    joint_df['closest_label'] = on_closest_labels
    joint_df['repnum'] = df.repnum.tolist()
    joint_df['roundnum'] = df['flyid'].str.split('_').str[-1]
    joint_df['fly'] = df.fly.tolist()
    joint_df['is_%s'%root_leg] = [True if root_leg in i else False for i in joint_df.closest_label]
    
    joint_df['is_contact'] = [True if i < dist_threshold else False for i in joint_df.distance]
    joint_df['interval'] = [int(i/interval) for i in joint_df.fnum]

    return joint_df

```
